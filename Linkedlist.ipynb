```python
class Node:
    def __init__(self,data=None, next=None):
        self.data = data
        self.next = next
class linkedlist:
    def __init__(self):
        self.obj = None
        
    def insert_at_beg(self,val):
        print("inserting "+str(val)+" to the linkedlist")
        node = Node(val,self.obj)
        self.obj = node
#         itr = self.obj.next
#         if itr is not None:
    
    def insert_at_end(self,val):
        itr = self.obj
        while itr.next:
            itr = itr.next
        node = Node(val)
        itr.next = node
    
    def del_at_beg(self):
        if self.obj is None:
            print("Linked list is empty")
            return
        temp = self.obj.next
        self.obj.next = None
        self.obj = temp
        print("\n")
        print("After deletion")
        self.printll()
        
    def traversal(self):
        print("\ntraversing the linkedlist")
        if self.obj is None:
            print("Linked list is empty")
            return
        itr = self.obj
        while itr:
            print(itr.data, end=",")
            itr = itr.next
            
    def printll(self):
        if self.obj is None:
            print("Linked list is empty")
            return
        itr = self.obj
#         print(itr.data)
        listll = ''+str(itr.data)
        while itr.next:
            listll+= "--->"+str(itr.next.data)
#             print(itr.next.data)
            itr = itr.next
        print(listll)
if __name__=="__main__":
    ll = linkedlist()
    ll.insert_at_beg(5)
    ll.insert_at_beg(15)
    ll.insert_at_beg(35)
    ll.printll()
    ll.insert_at_end(10)
    ll.traversal()
    ll.del_at_beg()
    ll.del_at_beg()
#     ll.printll()
```
